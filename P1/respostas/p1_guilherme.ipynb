## 4. Tratando os dados

```python
# Substituir zeros por NaN nas colunas selecionadas
cols_to_replace_zero = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol']

df_red[cols_to_replace_zero] = df_red[cols_to_replace_zero].replace(0, np.nan)
df_white[cols_to_replace_zero] = df_red[cols_to_replace_zero].replace(0, np.nan)

# Preencher os NaNs com a mediana de cada coluna
df_red[cols_to_replace_zero] = df_red[cols_to_replace_zero].fillna(df_red[cols_to_replace_zero].median())
df_white[cols_to_replace_zero] = df_white[cols_to_replace_zero].fillna(df_white[cols_to_replace_zero].median())
```

## 5. Filtrando pelos melhores vinhos tintos

```python
max_quality = df_red['quality'].max()
best_red_wines = df_red[df_red['quality'] == max_quality] 
best_red_wines
```

## 7. Distribuição de qualidade dos vinhos tintos

```python
plt.figure(figsize=(8, 5))
sns.histplot(df_red['quality'], bins=range(int(df_red['quality'].min()), int(df_red['quality'].max()) + 2), kde=False)
plt.title('Distribuição da Qualidade dos Vinhos')
plt.xlabel('Qualidade')
plt.ylabel('Frequência')
plt.show()
```

## 8. Distribuição de qualidade dos vinhos tintos

```python
plt.figure(figsize=(8, 5))
sns.histplot(df_white['quality'], bins=range(int(df_white['quality'].min()), int(df_white['quality'].max()) + 2), kde=False)
plt.title('Distribuição da Qualidade dos Vinhos')
plt.xlabel('Qualidade')
plt.ylabel('Frequência')
plt.show()
```

## 9. Análise de correlação de fatores para que um vinho tinto seja de boa qualidade

```python
# Calcular a matriz de correlação
correlation_matrix = df_red.corr()

# Exibir apenas as correlações com a coluna 'quality', ordenadas
quality_correlations = correlation_matrix['quality'].sort_values(ascending=False)
print("Correlação das variáveis com a qualidade do vinho:")
print(quality_correlations)

# Visualizar as correlações com um gráfico de barras para facilitar a interpretação
plt.figure(figsize=(8, 5))
sns.barplot(x=quality_correlations.index, y=quality_correlations.values)
plt.xticks(rotation=90)
plt.title('Correlação das Variáveis com a Qualidade do Vinho')
plt.ylabel('Coeficiente de Correlação')
plt.show()
```

#### *Conclusão da análise:* com os dados fornecidos podemos afirmar que existe uma grande correlação entre teor alcoólico com a qualidade do vinho. Vinhos com teor alcoólico
