## Informações sobre o Dataset

   Input variables (based on physicochemical tests):\
   1 - fixed acidity\
   2 - volatile acidity\
   3 - citric acid\
   4 - residual sugar\
   5 - chlorides\
   6 - free sulfur dioxide\

   8 - density\
   9 - pH\
   10 - sulphates\
   11 - alcohol
   
   Output variable (based on sensory data): 
   12 - quality (score between 0 and 10)

```python
import numpy as np
import pandas as pd
```

2.1 Exploração de Dados

```python
file_path_sample = 'dataset/winequality-red.csv'
column_names = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol','quality'] #Definição do nome das colunas
df_winered = pd.read_csv(file_path_sample, sep=';', names=column_names, low_memory=False)
df_winered.head()
```

```python
file_path_sample = 'dataset/winequality-white.csv'
column_names = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol','quality'] #Definição do nome das colunas
df_winewhite = pd.read_csv(file_path_sample, sep=';', names=column_names, low_memory=False)
df_winewhite.head()
```

2.2 Limpeza de Dados

```python
df_winered.isnull().sum().sort_values(ascending=False) #NÃO HÁ NULOS NEM NAN A SEREM TRATADOS
df_winered.isna().sum().sort_values(ascending=False)
```

```python
df_winewhite.isnull().sum().sort_values(ascending=False) #NÃO HÁ NULOS NEM NAN A SEREM TRATADOS
df_winewhite.isna().sum().sort_values(ascending=False)
```

```python
#Remover cabeçalho das linhas
df_winewhite = df_winewhite [1:]
df_winered = df_winered[1:]
```

2.3 Análise Exploratória (qualidade dos vinho, nível de acides e etc.)

Indicação dos melhores vinhos vermelhos

```python
quality_red_sort = df_winered.sort_values(by='quality', ascending=False)
quality_red_sort [:5]
```

Indicação dos melhores vinhos brancos

```python
quality_white_sort = df_winewhite.sort_values(by='quality', ascending=False)
quality_white_sort [:5]
```

Média de nota de qualidade para cada tipo de vinho

```python
winered = np.array(df_winered)
winewhite = np.array(df_winewhite)

quality_red = winered[:, 11].astype(int).mean()
quality_white = winewhite[:, 11].astype(int).mean()

print("A média de qualidade dos vinhos vermelhos é de:", quality_red)
print ("A média de qualidade dos vinhos brancos é de:",quality_white)
print ("A análise demonstra que o tipo de vinho não influencia na sua qualidade")
```

Média de sulfur nos top100 de cada vinho

```python
sulfur_red = winered[:100, 6].astype(float).mean()
sulfur_white = winewhite[:100, 6].astype(float).mean()

print("A média de sulfur nos melhores vinhos vermelhos é", sulfur_red)
print ("A média de sulfur nos melhores vinhos brancos é", sulfur_white)
print ("Os vinhos brancos de qualidade possuem maior sulfur total em sua composição")
```

PESQUISA DA INTERNET\
VINHOS BRANCOS SUAVES - PH 3,0–3,3\
VINHOS TINTOS SUAVES - PH 3,3–3,5

A quantidade de vinhos suaves por categoria

```python
suave_red = (df_winered ['pH'].astype(float)>=3.3) & (df_winered ['pH'].astype(float)<=3.5)
qnt_suavered = suave_red.sum()

suave_white = (df_winewhite ['pH'].astype(float)>=3.0) & (df_winewhite ['pH'].astype(float)<=3.3)
qnt_suavewhite = suave_white.sum()

print ("A quantidade de vinhos suaves tintos é:", qnt_suavered)
print ("A quantidade de vinhos suaves brancos é:", qnt_suavewhite)
print ("Na amostra coletada, há muito mais vinhos suaves brancos do que vinhos suaves tintos")
```

A quantidade de vinhos suaves de qualidade alta por categoria

```python
melhores_red5 = quality_red_sort [:5]
melhores_red_suave = (quality_red_sort ['pH'].astype(float)>=3.3) & (quality_red_sort ['pH'].astype(float)<=3.5)
qnt_suav_top5_red = melhores_red_suave.sum()

melhores_wihte5 = quality_white_sort [:5]
melhores_white_suave = (quality_white_sort ['pH'].astype(float)>=3.0) & (quality_white_sort ['pH'].astype(float)<=3.3)
qnt_suav_top5_white = melhores_white_suave.sum()

print ("A quantidade de vinhos suaves nos top 5 tintos é de:", qnt_suav_top5_red)
print ("A quantidade de vinhos suaves nos top 5 brancos é de:", qnt_suav_top5_white)
print ("Vinhos brancos dominam tanto na qualidade quanto na quantidade quando se trata do tipo suave")
```

Comparação em tabela dos melhores e piores vinhos brancos

```python
top5white = quality_white_sort [:5]
poor_quality_white_sort = df_winewhite.sort_values(by='quality', ascending=True)
low5white = poor_quality_white_sort [:5]

toplowwhite = pd.concat ([top5white,low5white], axis=0)
toplowwhite
```

Comparação em tabela dos melhores e piores vinhos tintos

```python
top5red = quality_red_sort [:5]
poor_quality_red_sort = df_winered.sort_values(by='quality', ascending=True)
low5red = poor_quality_red_sort [:5]

toplowred = pd.concat ([top5red,low5red], axis=0)
toplowred
```
