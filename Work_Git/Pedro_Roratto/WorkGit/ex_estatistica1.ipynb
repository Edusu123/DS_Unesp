```python
import os
import pandas as pd 
#get current working directory
cwd = os.getcwd()
print(cwd)
df = pd.read_csv(cwd + '/World Energy Consumption.csv')


#remove NaN
df = df.dropna()

#remove duplicates
df = df.drop_duplicates()
```
