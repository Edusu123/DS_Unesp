```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt
import os
os.getcwd()
```

```python
df = pd.read_csv('c:\\Users\\guino\\OneDrive\\Documents\\Unesp\\projects\\DS_Unesp\\Datasets\\lanchonete.csv')

df
```

```python
# Calcular o tempo total em minutos e selecionar apenas as colunas 'id' e 'tempo_total_minutos'
df_transformado = df.assign(tempo_total_minutos=df['Minutos'] + df['Segundos'] / 60).round(2)[['Entrega_ID', 'tempo_total_minutos']]

# Definir a mÃ©dia esperada foi dada (H0: mÃ©dia = 30 horas)
media_esperada = 30

# Temos que a mÃ©dia observada em Ha = 33.29
df_transformado['tempo_total_minutos'].mean()

# Realizar o teste t para amostras Ãºnicas
t_stat, p_val = stats.ttest_1samp(df_transformado['tempo_total_minutos'], media_esperada)

# Exibir os resultados
print(f"EstatÃ­stica t: {t_stat}")
print(f"p-valor: {p_val}")

# InterpretaÃ§Ã£o do p-valor
alpha = 0.05 # Escolhido por nÃ³s
if p_val < alpha:
    print("Rejeitamos a hipÃ³tese nula (Hâ). A mÃ©dia das horas trabalhadas Ã© significativamente diferente de 40 horas.")
else:
    print("Falhamos em rejeitar a hipÃ³tese nula (Hâ). A mÃ©dia das horas trabalhadas nÃ£o Ã© significativamente diferente de 40 horas.")
```
