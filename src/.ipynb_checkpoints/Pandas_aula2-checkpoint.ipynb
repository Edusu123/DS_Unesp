```python
# Criar uma função de categorização

def categoriza(inv):
    if inv >= 1500:
        return 'Big'
    elif (inv >= 600) and (inv < 1500):
        return 'Mediun'
    elif inv < 600:
        return 'Small'
```

```python
# chamando o método

df['cat_size'] = df['Area'].apply(categoriza)
df.head()
```

```python
# tamanhos categorizados
df.cat_size.value_counts()
```

```python
df.loc[df['Bairro'] == 'Botafogo']
```

## Utilizando Groupby
link: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.groupby.html

```python
# O mesmo resultado que o loc

botafogo = df.groupby("Bairro").get_group('Botafogo')
```

```python
botafogo
```

```python
df.groupby("cat_size").get_group('Big')[:10]
```

```python
df.groupby('cat_size').last()
```

```python
df.groupby('cat_size').first()
```

```python
# Agrupando por ID

df.groupby('cat_size').groups
```

```python
# Agrupando por size

df.groupby('cat_size').size()
```

```python
# Agrupando categoria, condominio e o valor

df.groupby('cat_size').agg({'Condominio':'count',
                           'Valor':'mean'})
```

```python
# Agrupando Tipo, condominio e o valor

df.groupby('Tipo').agg({'Condominio':'count',
                           'Valor':'mean'})
```

```python
df.head()
```

```python
# agrupando um tipo para duas métricas

df.groupby('Condominio').agg(['sum'])
```

```python
df.head()
```

```python
df.groupby('Bairro')[['IPTU','Valor']].sum()
```

```python
# agrupamento po tipo e bairro

df.groupby(by=df['Tipo']).Bairro.value_counts()[:10]
```

```python
grouped = df[['Valor','Bairro']].groupby(df['Tipo'])
grouped.mean()
```

```python
%time
grouped.Bairro.describe()
```

```python
df2 = pd.read_csv('aluguel2.csv', sep=';')
df2.head()
```

```python
# Valor bruto mensal para um inquilino

df2['Val_mes'] = df2['Valor'] + df2['Condominio'] + df2['IPTU']
df2.head()
```

```python
# Valor do metro quadrado

df2['Val_mq'] = (df2['Valor'] / df2['Area']).round(2)
df2.head()
```

```python
df2.Tipo.value_counts()
```

```python
# agrupando por tipo de residência

df2.groupby('Tipo').get_group('Casa')
```

```python
# Lista com os tipos de casa

casa = ['Casa', 'Casa de Condomínio', 'Casa de Vila']
```

```python
df2['Tipo_casa'] = df2['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento')
df2.head()
```

```python
df2.Tipo_casa.value_counts()
```

## Exercícios

### Problema: Análise de Dados de Vendas em uma Rede de Lojas

### Cenário:
Você é um cientista de dados trabalhando para uma rede de lojas de varejo. A empresa coletou dados de vendas de seus produtos em várias localidades durante o último ano. Seu objetivo é ajudar a empresa a entender melhor o desempenho das lojas, produtos mais vendidos e identificar tendências que podem auxiliar na criação de estratégias de vendas para o próximo ano.

### Análises:

 - Análise Exploratória:
    - Qual é o total de vendas por produto?
    - Qual é a loja que vendeu mais produtos?
    - Qual categoria de produto teve mais vendas em termos de quantidade?
    
 - Análise Temporal:
    - Identifique qual mês do ano teve mais vendas (quantidade) e qual mês gerou mais receita (valor total de vendas).
    
 - Agrupamento de Dados:
    - Agrupe os dados por categoria e calcule a média de vendas por produto dentro de cada categoria.
    - Quais produtos tiveram as melhores vendas em cada localidade?
    
 - Criação de Gráficos:
    - Crie gráficos de barras para visualizar as vendas por produto e por loja.
    - Crie um gráfico de linha para mostrar a evolução das vendas ao longo do tempo.
    
### Teremos que criar esse conjunto de dados com as seguinetes informações:

 - ID_Produto	
 - Nome_Produto	
 - Categoria
 - Preço
 - Quantidade_Vendida
 - ID_Loja
 - Localidade_Loja
